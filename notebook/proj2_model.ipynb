# **2. Basic Model**

```python
def set_params_basic(meal, Teletubby, dish, 
                     rating, nutrition, price, allergy_matrix, availability_matrix, cooking_time):
    # set up the parameters
    # rating
    r = rating.values.tolist()
    r = pulp.makeDict([Teletubby, dish], r, 0)
    
    # price and budget
    p = price['price'].values.tolist()
    p = pulp.makeDict([dish], p, 0)
    budget = 300
    price_budget = [p, budget]

    # allergy
    a = allergy_matrix
    a = pulp.makeDict([dish, dish], a, 0)

    # availability 
    z = availability_matrix
    z = pulp.makeDict([Teletubby, meal], z, 0)
    
    # nutrition
    n_sodium = nutrition['sodium'].tolist()
    n_sodium = pulp.makeDict([dish], n_sodium, 0)
    up_sodium = 2300 * 4 
    low_sodium = 0
    
    n_calory = nutrition['calories'].tolist()
    n_calory = pulp.makeDict([dish], n_calory, 0)
    up_calory = 2000 * 4 
    low_calory = 0

    n_fat = nutrition['fat'].tolist()
    n_fat = pulp.makeDict([dish], n_fat, 0)
    up_fat = 35 * 4 
    low_fat = 25 * 4 

    
    n_protein = nutrition['protein'].tolist()
    n_protein = pulp.makeDict([dish], n_protein, 0)
    up_protein = 46 * 4
    low_protein = 0
    
    
    nutri = {'sodium': [n_sodium, up_sodium, low_sodium],
             'calory': [n_calory, up_calory, low_calory],
             'fat': [n_fat, up_fat, low_fat],
             'protein': [n_protein, up_protein, low_protein]}
    
    # cooking time
    T = cooking_time['time'].values.tolist()
    T = pulp.makeDict([dish], T, 0)

    
    ls_params_basic = [r, nutri, price_budget, a, z, T]
    
    return ls_params_basic
```

```python
def basic_model(meal, Teletubby, dish, ls_params_basic):
    
    nm = len(meal)
    ns = len(Teletubby)
    nd = len(dish)
    
    # parameters
    r = ls_params_basic[0]
    nutri = ls_params_basic[1]
    price_budget = ls_params_basic[2]
    a = ls_params_basic[3]
    z = ls_params_basic[4]
    T = ls_params_basic[5]
    
    # set up the optimization problem
    meals = pulp.LpProblem("Meal_Plan", pulp.LpMaximize)
    
    # set the variable
    # d[j][k] --- whether dish j is selected in meal k
    d = pulp.LpVariable.dicts("dish", (dish, meal), 0, 1, pulp.LpInteger)

    # y[i][k] --- whether Teletubby i cooks in meal k
    y = pulp.LpVariable.dicts("cook", (Teletubby, meal), 0, 1, pulp.LpInteger)
    
    
    # objective function
    meals += (pulp.lpSum([r[i][j] * d[j][k] for i in Teletubby for k in meal for j in dish]), 'sum_of_rating')
    
    # constraits
    # the number of dishes in every meal --- no more than 3 dishes per meal 
    for k in meal:
        meals += (pulp.lpSum([d[j][k] for j in dish]) <= 3)

    # we eat the same dish at most twice
    for j in dish:
        meals += (pulp.lpSum([d[j][k] for k in meal]) <= 2)
        
    # nutrition
    for k in range(5):
        # sodium
        meals += (
            pulp.lpSum([nutri['sodium'][0][j] * d[j][k] for j in dish]) <= nutri['sodium'][1]
        )
        meals += (
            pulp.lpSum([nutri['sodium'][0][j] * d[j][k] for j in dish]) >= nutri['sodium'][2]
        )

        # calory
        meals += (
            pulp.lpSum([nutri['calory'][0][j] * d[j][k] for j in dish]) <= nutri['calory'][1]
        )
        meals += (
            pulp.lpSum([nutri['calory'][0][j] * d[j][k] for j in dish]) >= nutri['calory'][2]
        )

        # fat
        meals += (pulp.lpSum([nutri['fat'][0][j] * d[j][k] for j in dish]) <= nutri['fat'][1])
        meals += (pulp.lpSum([nutri['fat'][0][j] * d[j][k] for j in dish]) >= nutri['fat'][2])

        # protein
        meals += (
            pulp.lpSum([nutri['protein'][0][j] * d[j][k] for j in dish]) <= nutri['protein'][1]
        )
        meals += (
            pulp.lpSum([nutri['protein'][0][j] * d[j][k] for j in dish]) >= nutri['protein'][2]
        )
        
    # budget and price
    meals += (pulp.lpSum([price_budget[0][j] * d[j][k] for j in dish for k in meal]) <= price_budget[1])

    # allergy
    for k in meal:
        for j in dish:
            meals += (a[j][j] * d[j][k] == d[j][k])

    # cooking times constraints of y with z&T:
    # the Teletubbies who cook at meal k must be available
    for i in Teletubby:
        for k in meal:
            meals += (y[i][k] <= z[i][k])

    # each meal will be cooked by no more than 2 Teletubbies
    for k in meal:
        meals += (pulp.lpSum([y[i][k] for i in Teletubby]) <= 2)

    # the Teletubbies will cook 2~3 meals in total
    for i in Teletubby:
        meals += (pulp.lpSum([y[i][k] for k in meal]) <= 3)
        meals += (pulp.lpSum([y[i][k] for k in meal]) >= 2)

    # the minimum required cooking time of meal k 
    # should be no more than total available cooking time of all Teletubbies who cook for meal k
    for k in meal:
        for j in dish:
            meals += (T[j] * d[j][k] <= pulp.lpSum([y[i][k] * z[i][k] for i in Teletubby]))

    # solve the problem
    status = meals.solve()
    
    return meals
```

```python
def show_result(model, meal):
    # the optimizaiton objective function value is printed to the screen    
    print("Total Rating Score = ", pulp.value(model.objective))
    print('----------------------------')
    print('-------------------------------------------------------------------\n')

    # print the dishes selected for each meal and Teletubbies who cook for each meal
    weekday = {0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday'}
    for k in meal:
        # print weekday
        w = weekday[k]
        print(w + ': ')

        # print cooking Teletubby and menu
        ls_cook = []
        ls_dish = []
        for v in model.variables():
            if (str(v).startswith('cook')) & (str(v).endswith(str(k))) & (v.varValue == 1):
                ls_cook.append(str(v)[:-2][5:])
            if (str(v).startswith('dish')) & (str(v).endswith(str(k))) & (v.varValue == 1):
                ls_dish.append(str(v)[:-3][5:])

        if ls_dish == []:
            if ls_cook == []:
                # if all Teletubbies have class during the cooking time, eat outside
                print('No one is available! Eat outside!') 
            else:
                # if all Teletubbies who is available can not cook the dishes selected
                print('The meal does not meet the nutrition requirements! Eat outside!') 
        else:
            print('Teletubbies who cook: ' + str(ls_cook))
            print('Dishes: ')
            for i in ls_dish:
                print(i)

        print('-------------------------------------------------------------------\n')
```

```python
# set parameters
ls_params_basic = set_params_basic(meal, Teletubby, dish, 
                                   rating, nutrition2, price, allergy_matrix, availability_matrix, cooking_time)
```

```python
# solve the problem
meals_basic = basic_model(meal, Teletubby, dish, ls_params_basic)
```

```python
# show whether the problem is feasible
pulp.LpStatus[meals_basic.status]
```

```python
# show the optimal rating scores and weekly meal plan menu
show_result(meals_basic, meal)
```

#### **b. Solution to the potential problem**
To deal with this problem, we designed a two-stage model
* Stage-1: Get the cooking plan for other available days and output “Eat Outside” for the day no one is available.
* Stage-2: Make recommendations for eating outside.

```python
def eatoutside_stage1_params(meal, Teletubby, dish, 
                             rating, nutrition, price, allergy_matrix, availability_matrix, assistant_nutrition,
                             cooking_time, cooking_ability_matrix):
    # set up the parameters
    # rating
    r = rating.values.tolist()
    r = pulp.makeDict([Teletubby, dish], r, 0)
    
    # price and budget
    p = price['price'].values.tolist()
    p = pulp.makeDict([dish], p, 0)
    budget = 300
    price_budget = [p, budget]

    # allergy
    a = allergy_matrix
    a = pulp.makeDict([dish, dish], a, 0)

    # availability 
    z = availability_matrix
    z = pulp.makeDict([Teletubby, meal], z, 0)
    
    # nutrition
    n_sodium = nutrition['sodium'].tolist()
    n_sodium = pulp.makeDict([dish], n_sodium, 0)
    up_sodium = 2300 * 4 
    low_sodium = 0
    sodium = [0 for i in range(len(meal))]
    for idx in assistant_nutrition:
        sodium[idx] = low_sodium 
    sodium = pulp.makeDict([meal], sodium, 0)
    
    n_calory = nutrition['calories'].tolist()
    n_calory = pulp.makeDict([dish], n_calory, 0)
    up_calory = 2000 * 4 
    low_calory = 0
    calory = [0 for i in range(len(meal))]
    for idx in assistant_nutrition:
        calory[idx] = low_calory
    calory = pulp.makeDict([meal], calory, 0)

    n_fat = nutrition['fat'].tolist()
    n_fat = pulp.makeDict([dish], n_fat, 0)
    up_fat = 35 * 4 
    low_fat = 25 * 4 
    fat = [0 for i in range(len(meal))]
    for idx in assistant_nutrition:
        fat[idx] = low_fat 
    fat = pulp.makeDict([meal], fat, 0)

    n_protein = nutrition['protein'].tolist()
    n_protein = pulp.makeDict([dish], n_protein, 0)
    up_protein = 46 * 4
    low_protein = 0
    protein = [0 for i in range(len(meal))]
    for idx in assistant_nutrition:
        protein[idx] = low_protein 
    protein = pulp.makeDict([meal], protein, 0)
    
    nutri = {'sodium': [n_sodium, up_sodium, low_sodium, sodium],
             'calory': [n_calory, up_calory, low_calory, calory],
             'fat': [n_fat, up_fat, low_fat, fat],
             'protein': [n_protein, up_protein, low_protein, protein]}
    
    # cooking time
    T = cooking_time['time'].values.tolist()
    T = pulp.makeDict([dish], T, 0)

    # cooking ability --- M[i][j] means whehter student i can cook dish j
    M = cooking_ability_matrix
    M = pulp.makeDict([Teletubby, dish], M, 0)
    
    ls_params_eatoutside_stage1 = [r, nutri, price_budget, a, z, T, M]
    
    return ls_params_eatoutside_stage1
```

```python
def eatouside_model_stage1(meal, Teletubby, dish, ls_params_eatoutside_stage1):
    
    nm = len(meal)
    ns = len(Teletubby)
    nd = len(dish)
    
    # parameters
    r = ls_params_eatoutside_stage1[0]
    nutri = ls_params_eatoutside_stage1[1]
    price_budget = ls_params_eatoutside_stage1[2]
    a = ls_params_eatoutside_stage1[3]
    z = ls_params_eatoutside_stage1[4]
    T = ls_params_eatoutside_stage1[5]
    M = ls_params_eatoutside_stage1[6]
    
    # set up the optimization problem
    meals = pulp.LpProblem("Meal_Plan", pulp.LpMaximize)
    
    # set the variable
    # d[j][k] --- whether dish j is selected in meal k
    d = pulp.LpVariable.dicts("dish", (dish, meal), 0, 1, pulp.LpInteger)

    # y[i][k] --- whether Teletubby i cooks in meal k
    y = pulp.LpVariable.dicts("cook", (Teletubby, meal), 0, 1, pulp.LpInteger)
    
    
    # objective function
    meals += (pulp.lpSum([r[i][j] * d[j][k] for i in Teletubby for k in meal for j in dish]), 'sum_of_rating')
    
    # constraits
    # the number of dishes in every meal --- no more than 3 dishes per meal 
    for k in meal:
        meals += (pulp.lpSum([d[j][k] for j in dish]) <= 3)

    # we eat the same dish at most twice
    for j in dish:
        meals += (pulp.lpSum([d[j][k] for k in meal]) <= 2)
        
    # nutrition
    for k in range(5):
        # sodium
        meals += (
            pulp.lpSum([nutri['sodium'][0][j] * d[j][k] for j in dish]) + nutri['sodium'][3][k] <= nutri['sodium'][1]
        )
        meals += (
            pulp.lpSum([nutri['sodium'][0][j] * d[j][k] for j in dish]) + nutri['sodium'][3][k] >= nutri['sodium'][2]
        )

        # calory
        meals += (
            pulp.lpSum([nutri['calory'][0][j] * d[j][k] for j in dish]) + nutri['calory'][3][k] <= nutri['calory'][1]
        )
        meals += (
            pulp.lpSum([nutri['calory'][0][j] * d[j][k] for j in dish]) + nutri['calory'][3][k] >= nutri['calory'][2]
        )

        # fat
        meals += (pulp.lpSum([nutri['fat'][0][j] * d[j][k] for j in dish]) + nutri['fat'][3][k] <= nutri['fat'][1])
        meals += (pulp.lpSum([nutri['fat'][0][j] * d[j][k] for j in dish]) + nutri['fat'][3][k] >= nutri['fat'][2])

        # protein
        meals += (
            pulp.lpSum([nutri['protein'][0][j] * d[j][k] for j in dish]) + nutri['protein'][3][k]<=nutri['protein'][1]
        )
        meals += (
            pulp.lpSum([nutri['protein'][0][j] * d[j][k] for j in dish]) + nutri['protein'][3][k]>=nutri['protein'][2]
        )
        
    # budget and price
    meals += (pulp.lpSum([price_budget[0][j] * d[j][k] for j in dish for k in meal]) <= price_budget[1])

    # allergy
    for k in meal:
        for j in dish:
            meals += (a[j][j] * d[j][k] == d[j][k])

    # cooking times constraints of y with z&T:
    # the Teletubbies who cook at meal k must be available
    for i in Teletubby:
        for k in meal:
            meals += (y[i][k] <= z[i][k])

    # each meal will be cooked by no more than 2 Teletubbies
    for k in meal:
        meals += (pulp.lpSum([y[i][k] for i in Teletubby]) <= 2)

    # the Teletubbies will cook 2~3 meals in total
    for i in Teletubby:
        meals += (pulp.lpSum([y[i][k] for k in meal]) <= 3)
        meals += (pulp.lpSum([y[i][k] for k in meal]) >= 2)

    # the minimum required cooking time of meal k 
    # should be no more than total available cooking time of all Teletubbies who cook for meal k
    for k in meal:
        for j in dish:
            meals += (T[j] * d[j][k] <= pulp.lpSum([y[i][k] * z[i][k] for i in Teletubby]))

    # ability of cooking
    for k in meal:
        for j in dish:
            meals += (d[j][k] <= pulp.lpSum([y[i][k] * M[i][j] for i in Teletubby]))


    # solve the problem
    status = meals.solve()
    
    return meals
```

```python
def eatoutside_stage2_params(model, meal, Teletubby, dish, 
                             rating, nutrition, price, allergy_matrix):
    # get no_time_list: B
    B = [0 for i in range(len(meal))]
    C = [1 for i in range(len(dish))]
    spend = [0 for i in range(len(dish))]
    for v in model.variables():
        if (v.varValue == 1):
            
            # get "no_time_list": B
            if (str(v).startswith('cook')):
                B[int(str(v)[-1])] = 1
            
            # get "cook_dish_list": C
            # get "cook_dish_spend_list": spend
            if (str(v).startswith('dish')):
                dish_name = str(v)[:-3][5:].replace('_', ' ')
                idx = dish.index(dish_name + ' ')
                C[idx] = 0
                spend[idx] += price['price'].values.tolist()[idx]
    
    # no_time_list: B
    B = pulp.makeDict([meal], B, 0)
    
    # cook_dish_list: C
    C = pulp.makeDict([dish], C, 0)
    
    # price and new budget for eating outside
    p = price['price'].values.tolist()
    p = pulp.makeDict([dish], p, 0)
    eatoutside_budget = 300 - sum(spend)
    price_budget = [p, eatoutside_budget]
                
    # rating
    r = rating.values.tolist()
    r = pulp.makeDict([Teletubby, dish], r, 0)
    
    # allergy
    a = allergy_matrix
    a = pulp.makeDict([dish, dish], a, 0)
  
    # nutrition
    n_sodium = nutrition['sodium'].tolist()
    n_sodium = pulp.makeDict([dish], n_sodium, 0)
    up_sodium = 2300 * 4 
    low_sodium = 0
    sodium = [0 for i in range(len(meal))]
    for k in range(len(meal)):
        if B[k] == 1:
            sodium[k] = low_sodium 
    sodium = pulp.makeDict([meal], sodium, 0)
    
    n_calory = nutrition['calories'].tolist()
    n_calory = pulp.makeDict([dish], n_calory, 0)
    up_calory = 2000 * 4 
    low_calory = 0
    calory = [0 for i in range(len(meal))]
    for k in range(len(meal)):
        if B[k] == 1:
            calory[k] = low_calory
    calory = pulp.makeDict([meal], calory, 0)

    n_fat = nutrition['fat'].tolist()
    n_fat = pulp.makeDict([dish], n_fat, 0)
    up_fat = 35 * 4 
    low_fat = 25 * 4 
    fat = [0 for i in range(len(meal))]
    for k in range(len(meal)):
        if B[k] == 1:
            fat[k] = low_fat 
    fat = pulp.makeDict([meal], fat, 0)

    n_protein = nutrition['protein'].tolist()
    n_protein = pulp.makeDict([dish], n_protein, 0)
    up_protein = 46 * 4
    low_protein = 0
    protein = [0 for i in range(len(meal))]
    for k in range(len(meal)):
        if B[k] == 1:
            protein[k] = low_protein 
    protein = pulp.makeDict([meal], protein, 0)
    
    nutri = {'sodium': [n_sodium, up_sodium, low_sodium, sodium],
             'calory': [n_calory, up_calory, low_calory, calory],
             'fat': [n_fat, up_fat, low_fat, fat],
             'protein': [n_protein, up_protein, low_protein, protein]}
  
    ls_params_eatoutside_stage2 = [B, C, r, nutri, price_budget, a]
    
    return ls_params_eatoutside_stage2
```

```python
def eatouside_model_stage2(meal, Teletubby, dish, ls_params_eatoutside_stage2):
    
    nm = len(meal)
    ns = len(Teletubby)
    nd = len(dish)
    
    # parameters
    B = ls_params_eatoutside_stage2[0]
    C = ls_params_eatoutside_stage2[1]
    r = ls_params_eatoutside_stage2[2]
    nutri = ls_params_eatoutside_stage2[3]
    price_budget = ls_params_eatoutside_stage2[4]
    a = ls_params_eatoutside_stage2[5]
    
    # set up the optimization problem
    meals = pulp.LpProblem("Meal_Plan", pulp.LpMaximize)
    
    # set the variable
    # o[j][k] --- whether dish j is recommended to buy in meal k
    o = pulp.LpVariable.dicts("dish", (dish, meal), 0, 1, pulp.LpInteger)
 
    # objective function
    meals += (pulp.lpSum([r[i][j] * o[j][k] for i in Teletubby for k in meal for j in dish]), 'sum_of_rating')
    
    # constraits
    
    # If there exists someone who can cook meal k ---> do not buy dishes for this meal
    # If no one is available for cooking meal k ---> select 2-3 dishes and buy it in the restaurant
    for k in meal:
        meals += (pulp.lpSum([o[j][k] for j in dish]) * B[k] == 0)
        meals += (pulp.lpSum([o[j][k] for j in dish]) + B[k] <= 3)
        meals += (pulp.lpSum([o[j][k] for j in dish]) + B[k] >= 2)
    
    # do not buy dishes which has already been selected to cook
    for k in meal:
        for j in dish:
            meals += (o[j][k] <= C[j])
     
    # nutrition
    for k in range(5):
        # sodium
        meals += (
            pulp.lpSum([nutri['sodium'][0][j] * o[j][k] for j in dish]) + nutri['sodium'][3][k] <= nutri['sodium'][1]
        )
        meals += (
            pulp.lpSum([nutri['sodium'][0][j] * o[j][k] for j in dish]) + nutri['sodium'][3][k] >= nutri['sodium'][2]
        )

        # calory
        meals += (
            pulp.lpSum([nutri['calory'][0][j] * o[j][k] for j in dish]) + nutri['calory'][3][k] <= nutri['calory'][1]
        )
        meals += (
            pulp.lpSum([nutri['calory'][0][j] * o[j][k] for j in dish]) + nutri['calory'][3][k] >= nutri['calory'][2]
        )

        # fat
        meals += (pulp.lpSum([nutri['fat'][0][j] * o[j][k] for j in dish]) + nutri['fat'][3][k] <= nutri['fat'][1])
        meals += (pulp.lpSum([nutri['fat'][0][j] * o[j][k] for j in dish]) + nutri['fat'][3][k] >= nutri['fat'][2])

        # protein
        meals += (
            pulp.lpSum([nutri['protein'][0][j] * o[j][k] for j in dish]) + nutri['protein'][3][k]<=nutri['protein'][1]
        )
        meals += (
            pulp.lpSum([nutri['protein'][0][j] * o[j][k] for j in dish]) + nutri['protein'][3][k]>=nutri['protein'][2]
        )
        
    # budget and price
    meals += (pulp.lpSum([price_budget[0][j] * o[j][k] for j in dish for k in meal]) <= price_budget[1])

    # allergy
    for k in meal:
        for j in dish:
            meals += (a[j][j] * o[j][k] == o[j][k])

    # solve the problem
    status = meals.solve()
    
    return meals
```
