```python
import pandas as pd
import numpy as np

import json
import string
import re
```

```python
# cooking time
with open('gdrive/My Drive/project2/data/cooking_time.json', 'r') as ct:
    cooking_time = json.load(ct)

cooking_time = pd.DataFrame(cooking_time, index = ['time']).T
cooking_time['title'] = cooking_time.index
cooking_time.index = range(len(cooking_time['title']))
cooking_time
```

```python
# price 
with open('gdrive/My Drive/project2/data/recipe_prices.json', 'r') as pr:
    price = json.load(pr)

price = pd.DataFrame(price, index = ['price']).T
price['title'] = price.index
price.index = range(len(price['title']))
price
```

```python
# dish list
with open('gdrive/My Drive/project2/data/result.json', 'r') as f:
    data = json.load(f)
data = pd.DataFrame(data)
data
```

```python
data.isnull().sum()
```

```python
len(data[(data['title'].isnull() == True) | (data['title'].isnull() == True) | 
     (data['rating'].isnull() == True) | (data['calories'].isnull() == True) |
     (data['personal_rating'].isnull() == True) | (data['personal_rating'].isnull() == '[]')].index)
```

```python
data_nan = data.dropna()
data_nan
```

```python
data_nan.isnull().sum()
```

```python
data_merge_price = data_nan.merge(price, how = 'inner', on = 'title')
data_merge_price
```

```python
data_final = data_merge_price.merge(cooking_time, how = 'inner', on = 'title')
data_final
```
